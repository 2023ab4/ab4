```julia
import Ab4Paper2023
import AbstractTrees
import Makie
import CairoMakie
```

```julia
root_full = Ab4Paper2023.experiment_with_targets();
data_full = Ab4Paper2023.Data(root_full);
```

```julia
fig = Makie.Figure()
ax = Makie.Axis(fig[1,1], title="Initial library",
    width=700, height=400, yscale=log10, yminorticksvisible = true,
    yminorticks = Makie.IntervalsBetween(10),
    xgridvisible=false, ygridvisible=false,
    xlabel = "sequence", ylabel = "count",
    xticks = (0:10000:50000, ["0", "10000", "20000", "30000", "40000", "50000"]),
)
Makie.scatter!(ax, 1:length(data_full.counts[:,1]), sort(data_full.counts[:,1], rev=true), markersize=7)
#Makie.lines!(ax, 1:length(data_full.counts[:,1]), sort(data_full.counts[:,1], rev=true), color=:black)
Makie.xlims!(ax, -5000, 55000)
Makie.ylims!(ax, 10, 1e6)
Makie.resize_to_layout!(fig)
Makie.save("figures/initial_library_counts.pdf", fig)
fig
```
